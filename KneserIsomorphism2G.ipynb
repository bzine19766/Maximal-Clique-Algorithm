```python
import pickle

with open('coding.data', 'rb') as filehandle:
    # read the data as binary data stream
    placesList2 = pickle.load(filehandle)
```

```python
l=[]
for i in range(len(placesList2)):
    l.append(int(placesList2[i]))
```

```python
system=l
```

```python
z=[(e,list(filter(lambda x: x & e==0, system))) for e in system]
```

```python
H = nx.Graph()
```

```python
H.add_nodes_from(system)
```

```python
E=[]
```

```python
for i in range(len(z)):
    a,b =z[i]
    for e in b:
          E.append((a,e))
```

```python
E
```

```python
H.add_edges_from(E)
```

```python
neh=H.number_of_edges()
```

```python
neh
```

```python
nx.draw(H,node_size =1500,with_labels = True)
```

```python
nx.draw_shell(H,node_color='lightblue',node_size =1500,with_labels = True)
```

```python
nx.is_isomorphic(G, H)
```
